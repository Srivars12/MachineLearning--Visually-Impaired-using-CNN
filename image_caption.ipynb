```python
import pandas as pd
from PIL import Image
import IPython.display as display
import os

# Load the dataset
file_path = "ml_dataset.csv"  # Ensure the dataset path is correct
df = pd.read_csv(file_path)

# Set the image directory (Update this path to where your COCO images are stored)
image_directory = r"G:\coco-2017\coco2017\train2017"  # Update if needed

def get_caption(image_name):
    """Fetch captions for the given image filename."""
    captions = df[df["image_name"] == image_name]["caption"].tolist()
    return captions if captions else ["No caption found for this image."]

# User input
image_name = input("Enter the image filename (e.g., 000000203564.jpg): ")
captions = get_caption(image_name)

# Construct full image path
full_image_path = os.path.join(image_directory, image_name)

# Display Image and Captions
if os.path.exists(full_image_path):
    img = Image.open(full_image_path)
    display.display(img)
else:
    print("Image file not found. Make sure the image is in the correct directory.")

print("\nCaptions:")
for i, caption in enumerate(captions, 1):
    print(f"{i}. {caption}")
```
